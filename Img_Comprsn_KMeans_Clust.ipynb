```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
image=mpl.image.imread("dog.jpg")
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
X=image.reshape(675*1200,3) #or X=image.reshape(-1,3)
```

```python
X.shape
```

```python
kmeans=KMeans(n_clusters=10, n_init=10) #n_init tells how many times the image must go through the process of alloting centroids.
```

```python
kmeans.fit(X)
```

```python
segmented_img=kmeans.cluster_centers_[kmeans.labels_]
segmented_img=segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img/255)
```
